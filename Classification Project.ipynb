```python
x = df.iloc [:, 2:31]
y = df['diagnosis']
#print(x)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
la = LabelEncoder()
y = la.fit_transform(y)
print(y)
```

```python
from sklearn.model_selection import train_test_split
#from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = .20, random_state = 0)
```

```python
plt.plot(x_train, 'o', color = 'red')
plt.plot(x_test, 'o', color = 'black')
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
LRC = LogisticRegression ()
LRC.fit (x_train, y_train)

pred_lr = LRC.predict (x_test)
from sklearn.metrics import accuracy_score
accuracy_lr = accuracy_score(y_test, pred_lr)
print('Accuracy Score of Logistic Regression is : ', accuracy_lr)
print(pred_lr) 
```

```python
plt.plot(y_test, 'o', color = 'black')
plt.plot(pred_lr, 'o', color = 'red' )
```

```python
#plt.scatter(x_train, y_train, color = 'red')
```

```python
# K-Nearest Neighbours
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=15)

knn.fit(x_train,y_train)

#let us get the predictions using the classifier we had fit above
pred_knn = knn.predict(x_test)
print(pred_knn)

from sklearn.metrics import accuracy_score
accuracy_knn = accuracy_score(y_test, pred_knn)
print("Accuracy of K-Nearest Neighbours is : ", accuracy_knn)
```

```python
plt.plot(y_test, '^', color = 'black')
plt.plot(pred_knn, '^', color = 'red')
```

```python
# SVM 
from sklearn.svm import SVC
svc = SVC(kernel = 'rbf', C=1.0, random_state=0)
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
print(y_pred)

from sklearn import metrics
accuracy_svm = metrics.accuracy_score(y_test,y_pred)
print('Accuracy Score of svc with Kernel RBF is : ', accuracy_svm)
```

```python
# Decision Trees
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(criterion = "entropy", random_state = 100)
decision_tree.fit(x_train, y_train)
pred_dt =decision_tree.predict(x_test)
print(pred_dt)

from sklearn.metrics import accuracy_score
from math import sqrt
accuracy_dt = accuracy_score(y_test, pred_dt)
print('Accuracy Score of Decision Trees is : ',accuracy_dt)
```
